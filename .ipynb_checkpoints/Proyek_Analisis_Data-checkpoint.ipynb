### Cleaning Data

```python
hours_df["dteday"] = pd.to_datetime(hours_df["dteday"])
days_df["dteday"] = pd.to_datetime(days_df["dteday"])

hours_df.info()
print("")
days_df.info()
```

**Insight:**

## Exploratory Data Analysis (EDA)

### Menggabungkan data

```python
merged_df = pd.merge(hours_df, days_df.drop(columns=[
    'instant', 'yr', 'mnth', 'holiday', 'weekday',
    'workingday', 'weathersit', 'temp', 'atemp',
    'hum', 'windspeed', 'casual', 'registered', 'cnt']),
                     on='dteday', how='left')
```

```python
merged_df.info()
```

```python
merged_df.sample(5)
```

```python
merged_df.to_csv("merge_data.csv", index=False)
```

**Insight:**
- Data telah digabung dengan mengelompokkan data per hari dan melakukan agregasi untuk mendapatkan:
  1. Mean dari temp, atemp, hum, windspeed, dan weathersit.
  2. Sum dari casual, registered, cnt.

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
monthly_trend = merged_df.groupby(merged_df['dteday'].dt.to_period("M"))['cnt'].sum()

plt.figure(figsize=(12,6))
monthly_trend.plot(marker='o', linestyle='-', color='b')
plt.xticks(rotation=45)
plt.xlabel("Bulan")
plt.ylabel("Jumlah Penyewaan")
plt.title("Tren Penyewaan Sepeda per Bulan")
plt.grid()
plt.show()
```

**Insight:**
- Jumlah penyewaan sepeda meningkat drastis dari awal tahun sampai pertengahan tahun (Spring-Fall).
- Mengalami penurunan saat menjelang akhir tahun atau musim dingin (Winter).

### Pertanyaan 2:

```python
season_trend = merged_df.groupby("season_x")["cnt"].mean()

plt.figure(figsize=(8,5))
sns.barplot(x=season_trend.index, y=season_trend.values, palette="coolwarm", hue=season_trend.index, legend=False)
plt.xticks(ticks=[0, 1, 2, 3], labels=['Spring', 'Summer', 'Fall', 'Winter'])
plt.title("Rata-rata Penyewaan Sepeda per Musim")
plt.xlabel("Musim")
plt.ylabel("Jumlah Rata-rata Penyewaan Sepeda")
plt.grid(axis='y')
plt.show()
```

**Insight:**
- Peningkatan penyewaan dimulai pada musim panas (Summer) sampai musim gugur (Fall).
- Penurunan saat musim dingin (Winter).

### Pertanyaan 3:

```python
plt.figure(figsize=(10, 6))
sns.heatmap(merged_df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap Korelasi Faktor dengan Penyewaan Sepeda")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=merged_df['temp'], y=merged_df['cnt'], alpha=0.5)
plt.title("Scatter Plot: Suhu vs Penyewaan Sepeda")
plt.xlabel("Suhu")
plt.ylabel("Jumlah Penyewaan Sepeda")
plt.grid()
plt.show()
```

**Insight:**
- Faktor cuaca seperti temperatur, kelembaban, kecepatan angin, dan kondisi cuaca berperan penting dalam menentukan jumlah penyewaan sepeda. Semakin hangat suhu udara, semakin tinggi jumlah penyewaan sepeda.
- Penyewa terdaftar berdampak lebih besar dibandingkan dengan pengguna kasual dalam total penyewaan sepeda.

## Analisis Lanjutan (Opsional)

## Manual Grouping

### Group by hour of the day

```python
merged_df['time_of_day'] = pd.cut(merged_df['hr'], bins=[0, 6, 12, 18, 24], 
                                  labels=['Night', 'Morning', 'Afternoon', 'Evening'], right=False, ordered=False)
time_trend = merged_df.groupby('time_of_day', observed=False)['cnt'].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=time_trend.index, y=time_trend.values, palette="viridis", hue=time_trend.index, legend=False)
plt.title("Rata-rata Penyewaan Sepeda Berdasarkan Waktu dalam Sehari")
plt.xlabel("Waktu dalam Sehari")
plt.ylabel("Jumlah Rata-rata Penyewaan Sepeda")
plt.grid(axis='y')
plt.show()
```

**Insight:**
- Puncak penyewaan sepeda terjadi pada sore hari (Afternoon). Kemungkinan besar terjadi karena jam pulang kerja/sekolah, dimana banyak orang menggunakan sepeda untuk perjalanan pulang.
- Penurunan penyewaan sepeda terjadi pada malam hari (Night). Kemungkinan disebabkan oleh faktor keamanan, berkurangnya aktivitas luar ruangan, serta kondisi pencahayaan yang kurang mendukung untuk bersepeda.

### Group by weather situation

```python
weather_trend = merged_df.groupby("weathersit")["cnt"].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=weather_trend.index, y=weather_trend.values, palette="coolwarm", hue=weather_trend.index, legend=False)
plt.xticks(ticks=[0, 1, 2, 3], labels=["Clear", "Mist", "Light Snow/Rain", "Heavy Rain/Snow"])
plt.title("Rata-rata Penyewaan Sepeda Berdasarkan Kondisi Cuaca")
plt.xlabel("Kondisi Cuaca")
plt.ylabel("Jumlah Rata-rata Penyewaan Sepeda")
plt.grid(axis='y')
plt.show()
```

**Insight:**
- Puncak penyewaan sepeda terjadi ketika cuaca cerah (Clear).
- Penurunan penyewaan sepeda terjadi ketika cuaca buruk: hujan/bersalju (Heavy Rain/Snow).

### Group by season and user type

```python
season_user_trend = merged_df.groupby(['season_x', 'registered'], observed=False)['cnt'].mean().unstack()

fig, ax = plt.subplots(figsize=(10, 6))
season_user_trend.plot(kind='bar', stacked=True, colormap='viridis', ax=ax)
plt.title("Perbandingan Pengguna Terdaftar dan Kasual per Musim")
plt.xlabel("Musim")
plt.ylabel("Jumlah Rata-rata Penyewaan Sepeda")
plt.xticks(ticks=[0, 1, 2, 3], labels=["Spring", "Summer", "Fall", "Winter"], rotation=0)
plt.legend(["Casual", "Registered"], title="User Type", loc='upper left')
plt.grid(axis='y')
plt.show()
```

**Insight:**
- Musim gugur (Fall) memiliki jumlah penyewaan sepeda tertinggi.
- Pengguna terdaftar (Registered) jauh lebih dominan dibandingkan pengguna kasual (Casual) di setiap musim. Namun, pengguna kasual (Casual) cenderung lebih tinggi di musim panas dan musim gugur.
- Pada musim dingin, meskipun terjadi penurunan total penyewaan, pengguna terdaftar tetap cukup aktif.

## Conclusion

- Jumlah penyewaan sepeda meningkat drastis selama musim semi (Spring) hingga musim gugur (Fall).
- Musim gugur sebagai puncak penyewaan sepeda dan musim dingin menyebabkan penurunan penyewaan sepeda.
- Cuaca sangat berpengaruh dalam proses penyewaan sepeda, diikuti dengan waktu serta tipe pengguna. Semakin cuaca bagus maka semakin meningkat jumlah penyewaan sepeda, khususnya ketika sore hari, dimana para penyewa menggunakan sepeda sebagai alat transportasi. Selain itu, pengguna terdaftar sebagai penyewa dominan.
